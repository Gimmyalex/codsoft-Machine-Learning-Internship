# **Import libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score ,confusion_matrix
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
```

# **Loading data**

```python
data=pd.read_csv('/kaggle/input/bank-customer-churn-prediction/Churn_Modelling.csv')
data.head()
```

# **Data Preprocessing**

```python
data.info()
```

```python
data.drop(['RowNumber','CustomerId','Surname'],axis=1,inplace=True)
```

```python
data.duplicated().sum()
```

```python
plt.figure(figsize=(10,10))
sns.histplot(data=data,x='Age',kde=True)
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Age Distribution')
plt.show()
```

```python
gender_counts = data['Gender'].value_counts()
sns.barplot(x=gender_counts.index, y=gender_counts.values)
plt.title('Distribution of Male and Female Customers')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
data['Gender'] = data['Gender'].map({'Female': 1, 'Male': 0})
```

```python
new_geo=LabelEncoder()
data["new_geo"]=new_geo.fit_transform(data['Geography'])
data.drop('Geography',axis=1,inplace=True)
```

```python
data.head()
```

# **Feature Engineering**

```python
data['Balance_Salary_Ratio'] = data['Balance'] / data['EstimatedSalary']
data.drop(['Balance', 'EstimatedSalary'], axis=1, inplace=True)
```

```python
x=data.drop('Exited',axis=1)
y=data['Exited']
x
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

# **Model** **Selection**

```python
RFC=RandomForestClassifier(random_state=0)
RFC.fit(X_train,y_train)
y_pred1=RFC.predict(X_test)
```

# **Model Evaluation**

```python
print(accuracy_score(y_test,y_pred1))
print(confusion_matrix(y_test,y_pred1))
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(confusion_matrix(y_test,y_pred1),annot=True,fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
